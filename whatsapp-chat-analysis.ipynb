```python
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f=open('WhatsApp Chat with NRI Madarchods.txt','r',encoding='utf-8')
```

```python
data=f.read()
data
```

```python
# Define regex pattern
pattern = r"(\d{1,2}/\d{1,2}/\d{2}), (\d{1,2}:\d{2}\s?[APM\u202F\u00A0]{2,3}) - ([^:]+): (.*)"

# Find all matches
matches = re.findall(pattern, data)
matches
```

```python
# Convert matches into a DataFrame with correct column names
df = pd.DataFrame(matches, columns=["Date", "Time", "Sender", "Message"])

# Combine Date & Time into a single DateTime column
df["DateTime"] = pd.to_datetime(df["Date"] + " " + df["Time"], format="%m/%d/%y %I:%M %p")

# Reorder and drop old Date & Time columns
df = df[["DateTime", "Sender", "Message"]]

# Ensure proper 12-hour AM/PM formatting in final output
df["DateTime"] = df["DateTime"].dt.strftime("%m/%d/%y, %I:%M %p") 
# Display first few rows
print(df.head())
```

```python
# Convert DateTime back to datetime type
df["DateTime"] = pd.to_datetime(df["DateTime"], format="%m/%d/%y, %I:%M %p")

# Extracting date & time components
df['year'] = df['DateTime'].dt.year
df['month'] = df['DateTime'].dt.month_name()
df['day'] = df['DateTime'].dt.day
df['hour'] = df['DateTime'].dt.hour
df['minute'] = df['DateTime'].dt.minute

# Display first few rows
print(df.head())
```

```python
df['day_name']=df['DateTime'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
# df.head()
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour==23:
        period.append(str(hour)+'-'+str('00'))
    elif hour==0:
        period.append(str('00')+'-'+str(hour+1))
    else:
        period.append(str(hour)+'-'+str(hour+1))
```

```python
df['period']=period
```

```python
import seaborn as sns
```

```python
# Pivot table creation
heatmap_data = df.pivot_table(index='day_name', columns='period', values='Message', aggfunc='count').fillna(0)
print(heatmap_data)
sns.heatmap(heatmap_data)
plt.figure(figsize=(20,6))
plt.yticks(rotation=0)
plt.show()
```
